### 1. What does RGBA stand for?

RGBA stands for Red Green Blue Alpha. It is a color model used in computer graphics and digital imaging to represent colors with red, green, blue, and alpha (transparency) channels. Each channel is represented by an 8-bit value, ranging from 0 to 255, or a floating-point value ranging from 0.0 to 1.0. The combination of these four channels determines the color and transparency of a pixel. The red, green, and blue channels control the color intensity, while the alpha channel controls the transparency or opacity of the pixel.

### 3. What is a box tuple, and how does it work?


the box tuple provides a convenient way to define and manipulate rectangular regions within an image using pixel coordinates
A box tuple is defined as (left, upper, right, lower), where:

left represents the x-coordinate of the leftmost edge of the box.
upper represents the y-coordinate of the topmost edge of the box.
right represents the x-coordinate of the rightmost edge of the box.
lower represents the y-coordinate of the bottommost edge of the box.

### 7. What module contains Pillowâ€™s shape-drawing code?

```python
from PIL import Image, ImageDraw
image = Image.new("RGB", (500, 500), "white")
draw = ImageDraw.Draw(image)

# Draw a line from (100, 100) to (400, 400) with red color
draw.line([(100, 100), (400, 400)], fill="red")

image.show()
```

```python
from PIL import Image, ImageDraw

# Create a new image with a white background
image = Image.new("RGB", (400, 400), "white")
draw = ImageDraw.Draw(image)

# Draw the house shape
draw.rectangle([(100, 200), (300, 400)], fill="lightblue")  # House body
draw.polygon([(100, 200), (200, 100), (300, 200)], fill="orange")  # Roof

# Draw the door
draw.rectangle([(180, 300), (220, 400)], fill="brown")

# Draw the windows
draw.rectangle([(130, 250), (170, 290)], fill="white")  # Window 1
draw.rectangle([(230, 250), (270, 290)], fill="white")  # Window 2

# Display the image
# image.show()
image.save("house.jpg")
```
